## 3. Create a correlation Matrix heatmap

```python
# 3.1 Create a subplot with matplotlib just to show the subplots.

f,ax = plt.subplots(figsize=(10,10))
```

```python
# 3.2 Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

```python
corr.figure
```

The variables with stronger correlation are those with lighter colours. "asissts-points_per_game" and "rebounds-points_per_game" show that those players who are better scorers are also better rebounders and give more assists. The conclussion is that better players are norally better at more than one category.

## 4. Create scatterplots for strongest correlations

```python
# 4. Create a scatterplot for the variable with strongest corelation(1). "assists-points_per_game"

sns.lmplot(x = 'assists', y = 'points_per_game', data = df)
```

```python
# 4. Create a scatterplot for the variable with strongest corelation(2). "rebounds-points_per_game"

sns.lmplot(x = 'rebounds', y = 'points_per_game', data = df)
```

Both case have a strong positive correlation. We can see on both cases that as the production in points increases so does the production in assists or rebounds.

These two scatterplots also help us to identify outliers. We can see that some players can be great scorers but very average rebounders or assisters or viceversa. A player can be great at one part of the game but just average or worse at another.

As "assists" & "points per game" and "rebounds" & "points per game" have a strong correlation we could asssume that "assists" & "rebounds" have a strong correlation too however, that is not the case as we can see in the heat map (0.22) and the following scatter plot.

```python
# 4. (bonus)Create a scatterplot for the variables "rebounds-assists"

sns.lmplot(x = 'rebounds', y = 'assists', data = df)
```

We can see here lots of players with high numbers in rebounding but low in assists and viceversa. The conclussion is that not many players can be good at three or more aspects of the game.
